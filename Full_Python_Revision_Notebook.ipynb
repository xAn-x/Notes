### ```Linear Regression from scratch in numpy```

````python
w,b=np.random.rand(1),np.random.rand(1)
def model(x,w,b):
  return w*x+b

def loss(y,y_hat):
  diff=y_hat-y
  return np.mean(diff**2)


y_hat=model(x,w,b)

plt.plot(x,y,"b.")
plt.plot(x,y_hat,"r*")
plt.title("Without training...")
plt.text(-10,80,f"INITIAL-LOSS:{loss(y,y_hat)}",c="green")
````

````python
def train(epochs,x,y,w,b,alpha=0.001):
  losses=[]
  for epoch in range(epochs):
    y_hat=model(x,w,b)
    loss_val=loss(y,y_hat)

    # gradient-decent
    diff=y_hat-y
    w-=alpha*2*np.mean(diff*x)
    b-=alpha*2*np.mean(diff)

    #logs
    if (epoch+1)%100==0:
      print(f"{epoch+1}/{epochs}:Loss({loss_val})")

    losses.append(loss_val)

  return w,b,losses
````

````python
w,b,losses=train(1500,x,y,w,b)
````

````python
y_hat=model(x,w,b)
plt.plot(x,y,"b.")
plt.plot(x,y_hat,"r-")
plt.title("After training...")
plt.text(-10,80,f"FINAL-LOSS:{loss(y,y_hat)}",c="green")
````

### ```Using Pytorch,No Module```

````python
w,b=pt.rand(1,requires_grad=True),pt.rand(1,requires_grad=True) # make require_grad=True for for who u need to backprop from

pt_x,pt_y=pt.from_numpy(x).float(),pt.from_numpy(y).float()

def pt_model(x,w,b):
  return w*x + b

def pt_loss(y,y_hat):
  diff=y_hat-y
  return pt.mean(diff**2,dim=-1)

y_hat=pt_model(pt_x,w,b)

plt.plot(x,y,"b.")
plt.plot(x,y_hat.detach(),"r*")
plt.title("Without training...")
plt.text(-10,80,f"INITIAL-LOSS:{pt_loss(pt_y,y_hat)}",c="green")
````

````python
def train(epochs,x,y,w,b,alpha=0.001):
  losses=[]
  for epoch in range(epochs):
    y_hat=pt_model(x,w,b)

    # calculate loss
    loss_val=pt_loss(y,y_hat)

    # gradient-decent
    loss_val.backward()

    # update wts and bias
    with pt.no_grad():
      w-=alpha*w.grad
      b-=alpha*b.grad

    # reset grad
    w.grad.zero_()
    b.grad.zero_()

    #logs
    if (epoch+1)%100==0:
      print(f"{epoch+1}/{epochs}:Loss({loss_val})")

    losses.append(loss_val.item())

  return w,b,losses
````

````python
w,b,losses=train(1500,pt_x,pt_y,w,b)
````

````python
y_hat=pt_model(pt_x,w,b)
plt.plot(x,y,"b.")
plt.plot(x,y_hat.detach(),"r-")
plt.title("After training...")
plt.text(-10,80,f"FINAL-LOSS:{pt_loss(pt_y,y_hat)}",c="green")
````

### ```Using pytorch optimizer and loss_fn```

````python
def train(epochs,x,y,w,b,alpha=0.001):
  losses=[]
  loss_fn=F.mse_loss
  optimizer=pt.optim.SGD([w,b],lr=alpha)
  for epoch in range(epochs):
    y_hat=pt_model(x,w,b)

    # calculate loss
    loss_val=loss_fn(y_hat,y)

    # gradient-decent
    loss_val.backward()

    # update wts and bias
    optimizer.step()

    # zero_grad
    optimizer.zero_grad()

    #logs
    if (epoch+1)%100==0:
      print(f"{epoch+1}/{epochs}:Loss({loss_val})")

    losses.append(loss_val.item())

  return w,b,losses
````

### `CNN in torch`

````python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision.datasets import MNIST
from torchvision.transforms import ToTensor,Normalize,Compose
from torch.utils.data import DataLoader
````

````python
transforms=Compose([ToTensor(),Normalize((0.5,),(0.5,))])
train_ds=MNIST(root="data/",train=True,download=True,transform=transforms)
test_ds=MNIST(root="data/",train=False,download=True,transform=transforms)
````

````python
train_dl=DataLoader(train_ds,batch_size=64,shuffle=True)
test_dl=DataLoader(test_ds,batch_size=128,shuffle=True)
````

````python
import matplotlib.pyplot as plt
from torchvision.utils import make_grid

def show_batch(dl):
  for images,labels in dl:
    # denorm images
    images=images*0.5+0.5
    fig,ax=plt.subplots(figsize=(12,6))
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(make_grid(images,nrow=16).permute(1,2,0))
    break
````

````python
show_batch(train_dl)
````

````python
class ANN(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1=nn.Flatten()
    self.fc2=nn.Linear(28*28,128)
    self.fc3=nn.Linear(128,64)
    self.fc4=nn.Linear(64,10)
    self.relu=nn.ReLU()

  def forward(self,x):
    x=self.fc1(x)
    x=self.relu(x)
    x=self.fc2(x)
    x=self.relu(x)
    x=self.fc3(x)
    x=self.relu(x)
    x=self.fc4(x)
    return x
````

````python
class CNN(nn.Module):
  def __init__(self):
    super().__init__()
    self.conv1=nn.Conv2d(1,16,kernel_size=3,stride=2,padding=1)
    self.conv2=nn.Conv2d(16,32,kernel_size=3,stride=2,padding=1)
    self.conv3=nn.Conv2d(32,64,kernel_size=3,stride=2,padding=1)
    self.flatten=nn.Flatten()
    self.fc1=nn.Linear(64*4*4,10)
    self.relu=nn.ReLU()

  def forward(self,x):
    x=self.relu(self.conv1(x))
    x=self.relu(self.conv2(x))
    x=self.relu(self.conv3(x))
    x=self.flatten(x)
    x=self.fc1(x)
    return x
````

````python
from sklearn.metrics import accuracy_score
from tqdm.notebook import tqdm

def fit(epochs,model,train_dl,test_dl,optimizer,loss_fn,device=torch.device('cuda') if torch.cuda.is_available() else torch.device("cpu"),steps=5):
  history={"train_loss":[],"val_loss":[],"train_acc":[],"val_acc":[]}
  model.to(device)
  print(f"DEVICE:{device}\n")
  losses,accs=[],[]
  for epoch in tqdm(range(epochs)):
    model.train()
    for images,labels in train_dl:
      images,labels=images.to(device),labels.to(device)

      optimizer.zero_grad()
      y_hat=model(images)
      loss=loss_fn(y_hat,labels)

      train_loss=loss.item()
      losses.append(train_loss)
      train_accuracy=accuracy_score(labels.cpu(),y_hat.argmax(dim=1).cpu())
      accs.append(train_accuracy)

      loss.backward()
      optimizer.step()

    history["train_loss"].append(torch.tensor(losses).mean().item())
    history["train_acc"].append(torch.tensor(accs).mean().item())

    losses,accs=[],[]
    with torch.no_grad():
      model.eval()
      for images,labels in test_dl:
        images,labels=images.to(device),labels.to(device)

        y_hat=model(images)
        loss=loss_fn(y_hat,labels)
        val_loss=loss.item()
        losses.append(val_loss)
        val_accuracy=accuracy_score(labels.cpu(),y_hat.argmax(dim=1).cpu())
        accs.append(val_accuracy)

    history["val_loss"].append(torch.tensor(losses).mean().item())
    history["val_acc"].append(torch.tensor(accs).mean().item())

    if (epoch+1)%steps==0:
      print(f'{epoch+1}/{epochs}:\nTrain-Loss:{train_loss:.4f}, Train-Acc:{train_accuracy:.4f}\nVal-Loss:{val_loss:.4f}, Val-Acc:{val_accuracy:.4f}\n\n')

  return history



def plot_metric(history,metric):
  plt.plot(history[f'train_{metric}'],c="orange",marker='*',label=f"train_{metric}")
  plt.plot(history[f'val_{metric}'],c="blue",marker='o',label=f"val_{metric}")
  plt.title(metric,c="blue")
  plt.legend()
````

````python
ann_model=ANN()
cnn_model=CNN()
````

## Linear Regression from scratch:

````python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
````

````python
f_in=1

X=tf.random.normal(shape=(100,f_in))
y=2*X+3+tf.random.normal(shape=(100,f_in))

plt.plot(X,y,'b.')
````

````python
# Initialize weights and bias
wts = tf.Variable(tf.random.normal(shape=(f_in, 1)))
bias = tf.Variable(tf.random.normal(shape=(1,)))

print(f"wts: {wts}")
print(f"bias: {bias}")
````

````python
# Define the model function
def model(x, wts, bias):
    return tf.matmul(x, wts) + bias

# Training parameters
epochs, lr = 1000, 1e-2
losses = []

for epoch in range(epochs):
    with tf.GradientTape() as tape:
      y_hat = model(X, wts, bias)
      loss = tf.reduce_mean(tf.square(y - y_hat))

    # Compute gradients
    dw, db = tape.gradient(loss, [wts, bias])

    # Update weights and bias
    wts.assign_sub(lr * dw)
    bias.assign_sub(lr * db)

    losses.append(loss.numpy())

    # Print loss every 10 epochs
    if (epoch + 1) % 100 == 0:
        print(f"{epoch + 1}/{epochs}: Loss({loss.numpy()})")
        # break
````

````python
y_hat=model(X,wts,bias)

y_hat.shape,X.shape
````

````python
plt.plot(X,y,"b.")
plt.plot(X,model(X,wts,bias).numpy(),"r-")
````

### `Functional Modelling`

````python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, Dense, Flatten, MaxPooling2D, Input

# Define the input layer
input_layer = Input(shape=(28, 28, 1))

# Define the convolutional layers and pooling layers
x = Conv2D(16, 3, padding="valid", strides=1, activation="relu")(input_layer)
x = MaxPooling2D()(x)
x = Conv2D(32, 3, padding="same", strides=1, activation="relu")(x)
x = MaxPooling2D()(x)
x = Conv2D(64, 3, padding="valid", strides=1, activation="relu")(x)
x = MaxPooling2D()(x)

# Flatten the output and define the dense layer
x = Flatten()(x)
output_layer = Dense(10, activation="softmax")(x)

# Create the model
model = tf.keras.Model(inputs=input_layer, outputs=output_layer)

# Display the summary of the model
model.summary()
````

### `Model Subclassing`

````python
# Custom Layer
class ConvBlock(tf.keras.layers.Layer):
  def __init__(self,filters,kernel_size=3,strides=1,padding="valid",*args,**kwargs):
    super().__init__()
    self.conv_block=Sequential([
        Conv2D(filters,kernel_size,padding=padding,strides=strides,activation="relu",*args,**kwargs),
        MaxPooling2D()
    ])

  def call(self,x):
    return self.conv_block(x)


class CustomModel(tf.keras.models.Model):
  def __init__(self,img_shape=(28,28,1),num_classes=10):
    super().__init__()
    self.input_layer=InputLayer(input_shape=img_shape)
    self.conv_block_1=ConvBlock(16)
    self.conv_block_2=ConvBlock(32,padding="same")
    self.conv_block_3=ConvBlock(64)
    self.flatten=Flatten()
    self.dense=Dense(num_classes,activation="softmax")
    self.build((None,*img_shape)) # to specify the input shape

  def call(self,x):
    x=self.input_layer(x)
    x=self.conv_block_1(x)
    x=self.conv_block_2(x)
    x=self.conv_block_3(x)
    x=self.flatten(x)
    return self.dense(x)


model=CustomModel()
model.summary()
````

## `From pytorch to fastai`

````python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader


import torchvision
from torchvision import datasets
from torchvision.transforms import Compose,ToTensor,Lambda
````

````python
tfms=Compose([ToTensor(),Lambda(lambda x:x/255.)])

train_ds=datasets.FashionMNIST("/data/.",train=True,download=True,transform=tfms)
test_ds=datasets.FashionMNIST("/data/.",train=False,transform=tfms)
````

````python
train_dl=DataLoader(train_ds,batch_size=64,shuffle=True,num_workers=2,pin_memory=True)
test_dl=DataLoader(test_ds,batch_size=128,shuffle=True,num_workers=2,pin_memory=True)

len(train_dl),len(test_dl)
````

````python
for images,labels in train_dl:
  print(f"Images:{images.shape}")
  print(f"Labels:{labels.shape}")
  break
````

````python
from fastai.data.core import DataLoaders as fai_dls

dls=fai_dls.from_dsets(train_ds,test_ds,batch_size=63,num_workers=2)
````

````python
len(dls.train),len(dls.valid)
````

````python
dls.train_ds[0]
````

````python
# If your dataset doesn't have labels as strings, you need to add them:
dls.vocab = ['t-shirt/top', 'trouser', 'pullover', 'dress', 'coat',
             'sandal', 'shirt', 'sneaker', 'bag', 'ankle boot']
````

````python
dls.vocab
````

````python
class ConvBlock(nn.Module):
    def __init__(self, input_channels, output_channels, kernel_size=3, padding=0, stride=1):
        super().__init__()
        self.block = nn.Sequential(
            nn.Conv2d(input_channels, output_channels, kernel_size, padding=padding, stride=stride),
            nn.GELU(),
            nn.MaxPool2d(2,stride=2)
        )

    def forward(self, x):
        return self.block(x)


class FashionMNISTModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            ConvBlock(1, 32),
            ConvBlock(32, 64,5),
            ConvBlock(64, 128),
            nn.Flatten(),
            nn.Linear(128,100),
            nn.Linear(100,10)
        )

    def forward(self, x):
        return self.model(x)
````

````python
model=FashionMNISTModel()

model
````

````python
for images,labels in train_dl:
  print(model(images).shape)
  break
````

````python
from fastai.learner import Learner
from fastai.optimizer import OptimWrapper
import fastai.callback.schedule # for fit_one_cycle, lr_find
from fastai.metrics import accuracy, top_k_accuracy
from torch import optim
from torch.nn.functional import cross_entropy
from functools import partial

opt_func = OptimWrapper(params=model.parameters(),opt=optim.Adam)
top_3_accuracy = partial(top_k_accuracy, k=3,)

learn = Learner(dls, model, loss_func=cross_entropy, opt_func=opt_func, metrics=[accuracy,top_3_accuracy]).to_fp16()
````

````python
learn.lr_find()
````

````python
learn.fit_one_cycle(5,lr_max=slice(1e-2,1e-3))
````
